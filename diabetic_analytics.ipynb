```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the data present in the TSV file
df = pd.read_csv("diabetese.tsv", sep="\t")
df.head()
```

```python
# this info method helps us to disect the entire dataframe and provides the additional information
df.info()
```

```python
# mean is like finding the central tendency of the entire dataset
# mean is the the average of the data
```

```python
# Let's find the average age of diabetics patients
mean_age = df['AGE'].mean()
print(f"Mean age is : {mean_age}")
```

```python
# in some cases mean can give the misleading result 
# because of some outliers the entire mean of the dataset can get affected
# so we need to know that in what scenarios the mean can be helpful and where it can not
```

```python
# variance : how is the values are differing from mean
var_age = df['AGE'].var()
print(f"variance in age is :{var_age}")
```

```python
# Standard deviation is also like measuring variance 
# but it is more relatable as it gives result in that format only in which format our dataset is
```

```python
# let's find Standard deviation
std_age = df['AGE'].std()
print(f"Standard deviation is {std_age}")
```

```python
#BoxPlotting for the age field
plt.figure(figsize=(2,6))
# in boxplot we pass the array of values
plt.boxplot(df['AGE'], showmeans=True)
plt.grid(color='grey', linestyle='dotted')
plt.xlabel("Age")
plt.ylabel("Years")
plt.show()
```

```python
#Boxplotting for body mass index
plt.figure(figsize=(6,2))
plt.boxplot(df['BMI'], vert=False, showmeans= True)
plt.grid(color='gray', linestyle='dotted')
plt.xlabel("Value")
plt.ylabel('BMI')
plt.show()
```

```python
#Boxplotting for Blood Pressure
plt.figure(figsize=(6,2))
plt.boxplot(df['BP'], vert=False, showmeans= True)
plt.grid(color='gray', linestyle='dotted')
plt.xlabel("Value")
plt.ylabel('BP')
plt.show()
```

```python
# Histogram helps to identify the frequency of ocurance
# I want to understand what age of the people are mostly diabetics
# we can create histogram for the same 
```

```python
# let's create a histogram
df['AGE'].hist(bins=12)
plt.title('Age distribution of the diabetics patient')
plt.xlabel('Age Range')
plt.ylabel("Count")
plt.show()
```

```python
df['BMI'].hist(bins=12)
plt.title('BMI distribution of the diabetics patient')
plt.xlabel('BMI Range')
plt.ylabel("Count")
plt.show()
```
